```python
# Define the function
def replace_yes_no(df, columns):
    # Replace 'Yes' with 1 and 'No' with 0 for specified columns
    df[columns] = df[columns].replace({'Yes': 1, 'No': 0})
    return df
```

```python
# List of columns to modify
columns_to_change = ['owner_manager','female_owner', 'ceo_politician', 'top_manager_female', 'citizenship_top_manager', 'multinational_exp_top_manager', 'quality_certification', 'electrical_connection', 'power_outages',
'own_generator', 'application_water_connection', 'formal_planning', 'taxes_prevent_growth', 'labor_reg_prevent_growth', 'market_demand_prevent_growth', 'kpis', 'production_targets', 'performance_bonus',
'foreign_technology', 'informal_competitors', 'construction_permit', 'security', 'losses_fraudulent_transactions', 'savings_account', 'overdraft_facility', 'credit_line', 'outstanding_personal_loans', 'application_loans', 'certified_fin_statements', 'tax_inspection',
'online_tax_payment', 'government_contract', 'import_lic', 'operating_licence', 'vacancies']

# Call the function to replace 'Yes' and 'No' values
df = replace_yes_no(df, columns_to_change)
```
